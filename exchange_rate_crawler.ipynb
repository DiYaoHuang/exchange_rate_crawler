```python
import requests
import re
import datetime
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df = pd.DataFrame([])
date = datetime.datetime.now()
year = 2018
this_month = date.month
num_of_month = 3
for month in range(this_month, this_month-num_of_month, -1):
    url = 'https://rate.bot.com.tw/xrt/quote/{}-{}/USD'.format(year,'%02d'%month)
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'lxml')
    data = pd.read_html(soup.select('table')[0].prettify())[0]
    data.columns = ['掛牌日期','幣別','本行買入現金匯率','本行賣出現金匯率','本行買入即期匯率','本行賣出即期匯率','na']
    data = data.drop('na', axis=1)
    df = pd.concat([df,data])
    df = df.reset_index(drop=True)
df
```
